**Q1. Data Visualization and Statistical Measures:
For this question, you are required to analyse the iris dataset (iris.csv) using Python. Perform all
possible data visualization techniques (histograms, scatter plots, box plots, etc.) on all numerical
columns of the dataset. Additionally, calculate all possible statistical measures (mean, median,
mode, standard deviation, etc.) for each numerical column**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
import pandas as pd

iris_data = load_iris()
iris_df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
iris_df['species'] = iris_data.target
print(iris_df.head())
```

```python
summary_stats = iris_df.describe()
print("Summary Statistics:")
print(summary_stats)

iris_df.hist(figsize=(10, 8))
plt.suptitle("Histograms of Numerical Columns")
plt.show()

plt.figure(figsize=(10, 8))
sns.boxplot(data=iris_df)
plt.title("Box Plot of Numerical Columns")
plt.show()

plt.figure(figsize=(15, 10))
sns.pairplot(iris_df, diag_kind='kde')
plt.suptitle("Pairplot of Numerical Columns")
plt.show()
```

```python
mean_values = iris_df.mean()
print("Mean Values:")
print(mean_values)

median_values = iris_df.median()
print("\nMedian Values:")
print(median_values)

mode_values = iris_df.mode().iloc[0]
print("\nMode Values:")
print(mode_values)

std_dev_values = iris_df.std()
print("\nStandard Deviation Values:")
print(std_dev_values)

min_values = iris_df.min()
print("\nMinimum Values:")
print(min_values)

max_values = iris_df.max()
print("\nMaximum Values:")
print(max_values)
```
